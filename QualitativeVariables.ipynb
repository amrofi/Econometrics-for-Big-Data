```r
Wage <- data.frame(Wage)
cat("Column names...", names(Wage), "\n", sep="\n")
```

```r
#Linear regression of Wage on Education
lm_wage <- lm(wage~as.factor(education), data=Wage)
summary(lm_wage)
```

```r
#Linear regression of Wage on Education and Age
lm_wage1 <- lm(wage~as.factor(education)+age, data=Wage)
summary(lm_wage1)
```

```r
#Linear regression of Wage on Education ,Race, and Age
lm_wage2 <- lm(wage~as.factor(education)+age+as.factor(race), data=Wage)
summary(lm_wage2)
```

```r
Auto <- read.csv("auto.csv")
cat("Column names...", names(Auto), "\n", sep="\n")
```

```r
#calculating R^2 value for regression of mpg on year
Auto1 <- lm(mpg~year,data=Auto)
ggplot(Auto, aes(x = year, y = mpg)) +
geom_point() +
stat_smooth(method = "lm", col = "red") +
ggtitle("Linear regression of MPG on Year")
cat("R^2 = ", summary(Auto1)$r.squared)
```

```r
Auto2 <- lm(mpg~year+weight,data=Auto)
summary(Auto2)$r.squared
```

```r
Auto3 <- lm(mpg~year+weight+horsepower+displacement+cylinders,data=Auto)
summary(Auto3)$r.squared
```

```r
Auto4 <- lm(mpg~year+weight+horsepower+displacement+cylinders+as.factor(name),data=Auto)
summary(Auto4)$r.squared
```
