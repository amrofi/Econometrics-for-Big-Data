## Linear Regression

A function to show R-squares, slope, intercept and P value from the fit as the title of the plot

```r
RegressionData <- function (fit) {

  cat("\n", "Fit Data...")
  cat("\n", "Adj R2 = ",signif(summary(fit)$adj.r.squared, 5), "\n",
      "Intercept =",signif(fit$coef[[1]],5 ), "\n",
      "Slope =",signif(fit$coef[[2]], 5), "\n",
      "P =",signif(summary(fit)$coef[2,4], 5))
    
}
```

```r
#import table
library(readr)
library(ggplot2)

Auto <- read_csv("Auto.csv")

#reading column names
cat("Column names...", names(Auto), "\n", sep="\n")

#create arrays for mpg and weight column
mpg <- Auto$"mpg"
weight <- Auto$"weight"

#calculating the mean of mpg array
cat("The mean for MPG is...",mean(mpg), "\n", sep="\n")


#creating linear regression for mpg on weight (Y on X)
relation <- lm(mpg~weight)

#plotting data
ggplot(Auto, aes(x = weight, y = mpg), main='Linear Regression of MPG on Weight') + 
  geom_point() +
  stat_smooth(method = "lm", col = "red") +
  ggtitle('Linear Regression of MPG on Weight')

#Regression Data
RegressionData(relation)
```
