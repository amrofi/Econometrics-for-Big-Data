## Linear Regression

```r
#import table
library(readr)
Auto <- read_csv("Auto.csv")

#reading column names
cat("Column names...", names(Auto), "\n", sep="\n")

#create arrays for mpg and weight column
mpg <- Auto$"mpg"
weight <- Auto$"weight"

#calculating the mean of mpg array
cat("The mean for MPG is...",mean(mpg), "\n", sep="\n")


#creating linear regression for just mpg and mpg on
#weight respectively
relation0 <- lm(mpg~1)
relation1 <- lm(mpg~weight)
print(relation0)
print(relation1)

#plotting data
plot(weight, mpg, abline(lm(mpg~weight)) , col ="blue")
```
